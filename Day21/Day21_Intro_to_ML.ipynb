# Day 21 – Introduction to Machine Learning

Machine Learning is a way of teaching computers to learn patterns from data instead of giving them fixed rules. In traditional programming, we give data and instructions, and the computer produces results. In Machine Learning, we give data and correct answers, and the computer finds the rules by itself. These learned rules are then used to make predictions on new data.

```python
Machine Learning Problem Classification Notes:
1. Supervised Learning
Definition:
Supervised Learning is a type of machine learning where the data contains input-output pairs (labels).
The model learns to map inputs → correct outputs.

Real-World Examples:

| Problem                 | Input          | Output Label         |
| ----------------------- | -------------- | -------------------- |
| Spam Detection          | Email text     | Spam / Not Spam      |
| House Price Prediction  | House features | Price                |
| Disease Diagnosis       | Medical report | Disease / No Disease |
| Loan Default Prediction | Financial data | Default / No Default |

2.Unsupervised Learning
Definition:
Unsupervised Learning is when the data has no labels.
The model tries to discover hidden patterns, structures, or clusters.

Real-World Examples:

| Problem                 | Goal                                |
| ----------------------- | ----------------------------------- |
| Customer Segmentation   | Group similar customers             |
| Market Basket Analysis  | Find products bought together       |
| Topic Modeling          | Discover hidden topics in documents |
| Fraud Pattern Discovery | Detect unusual behavior             |

3.Reinforcement Learning
Definition:
Reinforcement Learning (RL) is when an agent learns by interacting with an environment and receiving rewards or penalties.
    
Real-World Examples:

| Problem          | Reward Signal                   |
| ---------------- | ------------------------------- |
| Chess AI         | Win (+), Lose (-)               |
| Robot Navigation | Reach goal (+), Crash (-)       |
| Self-Driving Car | Safe driving (+), Collision (-) |
| Game Playing AI  | Score points                    |

4.Quick Comparison Table:

| Feature           | Supervised                 | Unsupervised          | Reinforcement              |
| ----------------- | -------------------------- | --------------------- | -------------------------- |
| Labels Available? | ✅ Yes                      | ❌ No                  | ❌ No (reward instead)      |
| Goal              | Predict known output       | Discover structure    | Learn best actions         |
| Training Style    | Learn from correct answers | Find patterns         | Trial & error with rewards |
| Example           | Spam Detection             | Customer Segmentation | Game AI                    |

Step-by-Step Method to Classify Any Problem :
Step 1:
Ask → Do I have labeled data?
Yes → Supervised Learning
No → Go to Step 2

Step 2:
Ask → Am I trying to find hidden patterns or groups?
Yes → Unsupervised Learning
No → Go to Step 3

Step 3:
Ask → Is there an agent making decisions and receiving rewards over time?
Yes → Reinforcement Learning
No → Re-evaluate the problem
```
